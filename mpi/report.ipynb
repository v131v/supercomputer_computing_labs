```python
import matplotlib.pyplot as plt
```

# 1 lab

```python
#   
sizes = [100000, 1000000, 10000000]
times_4 = [0.000194788, 0.00188994, 0.00665903]
times_8 = [0.000118971, 0.000928879, 0.00373888]
times_16 = [9.89437e-05, 0.000637054, 0.00260305]

#  
plt.figure(figsize=(8, 6))
plt.plot(sizes, times_4, marker='o', label='4 processes', color='b')
plt.plot(sizes, times_8, marker='o', label='8 processes', color='g')
plt.plot(sizes, times_16, marker='o', label='16 processes', color='r')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Vector Size (log scale)')
plt.ylabel('Time Taken (log scale)')
plt.title('Time vs Vector Size for Different Number of Processes')
plt.legend()
plt.grid(True)

#  
plt.show()
```
