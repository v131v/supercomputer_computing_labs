```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1 lab

```python
# Данные из логов
sizes = [1_000_000, 10_000_000, 100_000_000, 1_000_000_000]
time_with_reduction = [0.00265002, 0.00641608, 0.0689671, 0.458701]
time_without_reduction = [0.000974894, 0.00454497, 0.0496259, 0.459511]

# Построение графиков
plt.figure(figsize=(10, 6))

plt.plot(sizes, time_with_reduction, label="With Reduction", marker="o", linestyle="-", color="blue")
plt.plot(sizes, time_without_reduction, label="Without Reduction", marker="o", linestyle="--", color="orange")

# Настройки графика
plt.title("Comparison of Execution Times with and without Reduction", fontsize=14)
plt.xlabel("Vector Size", fontsize=12)
plt.ylabel("Execution Time (seconds)", fontsize=12)
plt.xscale("log")
plt.yscale("log")
plt.xticks(sizes, [f"{size:,}" for size in sizes], rotation=45)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)

# Показать график
plt.tight_layout()
plt.show()

# Расчёт ускорения (speedup)
speedup = np.array(time_with_reduction) / np.array(time_without_reduction)

plt.figure(figsize=(10, 6))
plt.plot(sizes, speedup, marker="o", linestyle="-", color="green", label="Speedup")

# Настройки графика для ускорения
plt.title("Speedup with Reduction over Without Reduction", fontsize=14)
plt.xlabel("Vector Size", fontsize=12)
plt.ylabel("Speedup", fontsize=12)
plt.xscale("log")
plt.xticks(sizes, [f"{size:,}" for size in sizes], rotation=45)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.axhline(y=1, color="red", linestyle="--", linewidth=1, label="No Speedup")
plt.legend(fontsize=12)

# Показать график ускорения
plt.tight_layout()
plt.show()
```
